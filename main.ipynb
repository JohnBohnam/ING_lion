```python
import pandas as pd

# Load the two CSV files into pandas DataFrames
df1 = pd.read_csv('./data/target/imputed_data.csv')
df2 = pd.read_csv('./data/development_sample_transformed.csv')

df1.drop(columns=['Unnamed: 0'], inplace=True)

column_mapping = {
    "application_ammout": "application_amount",
    "installment_ammount": "installment_amount",
    
}

df1.rename(columns=column_mapping, inplace=True)

df2.drop(columns=["application_date"], inplace=True)


df2.head()
```

```python
df1.head()
```

```python


# Check if columns are the same
if df1.columns.equals(df2.columns):
    print("Columns are the same.")
else:
    print("Columns are not the same.")
    for (col1, col2) in zip(df1.columns, df2.columns):
        if col1 != col2:
            print(f"\"{col1}\" : \"{col2}\"")



# Initialize a list to store mismatch locations
mismatch_locations = []

# Iterate over each row in the DataFrames and compare corresponding values
for index, row in df1.iterrows():
    for column in df1.columns:
        if row[column] != df2.iloc[index][column]:
            mismatch_locations.append((index, column))

# Print mismatch locations
if mismatch_locations:
    print("Mismatch locations:")
    for index, column in mismatch_locations:
        print(f"Row {index} - Column {column}")
else:
    print("No mismatches found.")
```
